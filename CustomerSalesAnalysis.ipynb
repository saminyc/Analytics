### #5 frequency of items sold from inventory

```sql
USE AdventureWorks2017
select count(distinct(pth.ProductID)) as 'total_products_sold' 
from Production.TransactionHistory pth;
```

### #6 Order frequency for each product ordered, from highest to lowest

```sql
USE AdventureWorks2017
select pth.ProductID as product_id ,count(pth.TransactionID) as num_of_orders
from Production.TransactionHistory pth
GROUP BY ProductID
ORDER BY COUNT(PTH.TransactionID) DESC;
```

#7 P<span style="background-color: rgb(19, 19, 20); color: rgb(227, 227, 227); font-family: &quot;Google Sans&quot;, &quot;Helvetica Neue&quot;, sans-serif; font-size: 16px; white-space: pre-wrap; caret-color: rgb(227, 227, 227);">roduct information and the per unit selling cost of all products</span>

```sql
use AdventureWorks2017
select distinct(p.ProductID) as 'product_id',p.Name as 'product_name',s.UnitPrice as 'per_unit_cost'
from Production.Product p
FULL JOIN Sales.SalesOrderDetail as s on p.ProductID=s.ProductID
where s.UnitPrice IS NOT NULL;
```

### #8 <span style="color: rgb(227, 227, 227); font-family: &quot;Google Sans&quot;, &quot;Helvetica Neue&quot;, sans-serif; font-size: 16px; white-space: pre-wrap; caret-color: rgb(227, 227, 227);">The total number of orders placed by each customer from most to least</span>

```sql
use AdventureWorks2017
SELECT soh.CustomerID as customer_id,count(sod.ProductID) as num_of_orders 
from sales.SalesOrderDetail sod
FULL JOIN Sales.SalesOrderHeader soh on sod.SalesOrderID=soh.SalesOrderID
GROUP BY soh.CustomerID
ORDER BY num_of_orders DESC;
```

#9 <span style="color: rgb(227, 227, 227); font-family: &quot;Google Sans&quot;, &quot;Helvetica Neue&quot;, sans-serif; font-size: 16px; caret-color: rgb(227, 227, 227);">Top-selling colors, based on the order volume</span>

```sql
use AdventureWorks2017
select p.color as product_color,sum(s.OrderQty) as ordered_qty 
from Production.Product p
FULL JOIN Sales.SalesOrderDetail s on s.ProductID=p.ProductID
GROUP BY (p.Color)
ORDER BY ordered_qty DESC;
```
